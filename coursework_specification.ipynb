## Introduction to Ciphers

Ciphers have been used throughout history to transmit secret messages. Julius Caesar shifted letters to obscure military orders; during World War II, the Enigma machine encoded communications using systematic character substitution. A cipher transforms a secret message into an encoded form using a secret key; the intended recipient uses the same key to decode it.

### Cipher Alphabets

Each cipher scheme in this coursework operates on a *cipher alphabet* $\sigma = \sigma_0 \sigma_1 \ldots \sigma_{k-1}$: an ordered sequence of distinct lower-case alphabetical characters. The cipher alphabet is naturally a subset of the ambient alphabet. Here are a few named examples of cipher alphabet we will refer to throughout the coursework:

| name  | cipher alphabet | 
|--------|-----|
| `default` | `abcdefghijklmnopqrstuvwxyz` | 
| `vowels` | `aeiou` | 
| `consonants` | `bcdfghjklmnpqrstvwxyz` | 
| `qwerty` | `qwertyuiopasdfghjklzxcvbnm` | 
| `ambidextrous` | `ambidextrous` |

In addition to these, the coursework may specify more cipher alphabets in specific tasks.

In many cipher schemes, the indices of letters in the alphabet are required:

* **Indices start with `0`**, increasing from left to right, so that index `0` letter in `vowels` cipher is `a` and index `3` letter is `o`.
* Indices are assumed to **wrap around**, so that index `5` in `vowels` is again `a` and the index `7` is is `i`. 
* **Negative indices are allowed** (also wrapping around), so that index `-1` in `vowels` is `u` and index `-12` is `o`. 


### Substitution Ciphers

A **substitution cipher** replaces each cipher alphabet character according to an arbitrary bijection: every input character maps to a distinct output character, the mapping covers every cipher alphabet character, and no non-cipher-alphabet character is introduced. Thus, it suffices to specify how each index of the cipher alphabet is mapped to another index. Characters that are not in the cipher alphabet are left unchanged, with one exception:

**Capitalisation rule**: Wherever a cipher alphabet character appears in the input message in uppercase, its encoded counterpart is also capitalised.


For instance, if your substituion cipher replaces `agent shaken not stirred` by `agert wyaher rot wtilled`, it should replace `Agent 007: "Shaken, not stirred!"` by `Agert 007: "Wyaher, rot wtilled!"`. Note: Since `s` is replaced by `w`, `S` must be replaced by `W`, while numeric characters like `0`, `7`, symbols `,`, `"`, and `!` are unaffected (as are blank spaces ` `).

### Cipher Notation

The following notation is used throughout the coursework. Each cipher is parameterised
by a **cipher alphabet** $\sigma$; when $\sigma$ is omitted, the `default` alphabet is
assumed.

- $C_{c,\sigma}$ — Caesar cipher with shift $c$ on alphabet $\sigma$.
- $A_{a,b,\sigma}$ — Affine cipher with multiplier $a$ and offset $b$ on alphabet $\sigma$.
- $S_{p,\sigma}$ — Substitution cipher with permutation $p$ on alphabet $\sigma$.

Two named permutations are used throughout:

- $S_{\texttt{reverse}, \sigma}$: reverses the alphabet $\sigma$, $p(i) = k-1-i$.
- $S_{\texttt{split}, \sigma}$: places even-indexed letters (in order) in the alphabet $\sigma$ first, then odd-indexed letters
  (in reverse order). On the `default` alphabet this gives `acegikmoqsuwyzxvtrpnljhfdb`.

When $\sigma = \texttt{default}$, the $\sigma$ subscript can be dropped. For e.g. $C_c = C_{c,\texttt{default}}$.

## Part B: Cipher Manipulation and Composition &nbsp;&nbsp; **[8 marks]**



Let $R$, $R_1$, $R_2$ denote arbitary ciphers. The **composition**
$R_2 \circ R_1$ means applying $R_1$ first, then $R_2$:
$(R_2 \circ R_1)(x) = R_2(R_1(x))$.

The **power** $R^n$ for non-negative integer $n$ denotes composing $R$ with itself
$n$ times; $R^0$ is the identity cipher.

For Caesar ciphers on the **same** alphabet $\sigma$:
$$C_{c_1,\sigma} + C_{c_2,\sigma} = C_{c_1+c_2,\sigma}, \qquad
  nC_{c,\sigma} = C_{nc,\sigma} \quad (n \in \mathbb{Z})$$

For affine ciphers sharing the same multiplier $a$ on the **same** alphabet $\sigma$:
$$A_{a,b_1,\sigma} + A_{a,b_2,\sigma} = A_{a,\,b_1+b_2,\,\sigma}, \qquad
  nA_{a,b,\sigma} = A_{a,\,nb,\,\sigma} \quad (n \in \mathbb{Z})$$

**Important:** addition and scalar multiplication are only defined when both operands
share the same cipher alphabet. Adding ciphers with different alphabets should be forbidden.

## Part C: Affine Hill Cipher &nbsp;&nbsp; **[4 marks]**



The **affine Hill cipher** generalises the affine cipher to operate on *blocks* of $n$
characters at a time, using matrix multiplication instead of scalar multiplication.


### Block Ciphers

As with substitution ciphers, in a block cipher, characters that are not in the cipher alphabet $\sigma$ are unmodified and capitalisation of characters is respected by location. 

However, unlike a substituion cipher, the cipher-alphabet characters are not substituted indvidually and independently. Instead, we form blocks of $n$ consecutive cipher-alphabet characters appearing in the secret message and find their respective substitutions one block at a time. 

Given a length $k$ cipher-alphabet $\sigma$, a block of $n$ consecutive cipher-alphabet characters in the secret message is $\mathbf{c}=\sigma_{x_1}\sigma_{x_2}..\sigma_{x_{n}}$, defined by the vector $\mathbf{x}=(x_1, x_2, \ldots, x_n)$, where $x_i$ are (possibly non-unique) numbers in the set $\{0,1,\ldots, k-1\}$.

The task in a block cipher is to map each block $\mathbf{c}$ to another block $\mathbf{d}=\sigma_{y_1}\sigma_{y_2}..\sigma_{y_n}$, following some logic, or equivalently, mapping the vector of  cipher-alphabet indices $\mathbf{x}=(x_1, x_2, \ldots, x_n)$ to another, $\mathbf{y}=(y_1, y_2, \ldots, y_n)$. 

### Task C1: Affine Hill Cipher &nbsp;&nbsp; [2 marks]

The **affine Hill cipher**
$H_{A,\mathbf{b},\sigma}$ defined by an $n \times n$ integer matrix $A$ and an integer
vector $\mathbf{b}$ of length $n$ is a block cipher with block size $n$, acting on cipher
alphabet $\sigma$. Given the vector $\mathbf{x}$ representing a block $\mathbf{c}$, it
computes the replacement block $\mathbf{d}$ via:

$$ y_i = A_{i,1}x_1 + A_{i,2}x_2 + \cdots + A_{i,n}x_n + b_i, \qquad i = 1, \ldots, n,$$

with cipher-alphabet indices wrapping around as required.

**Example.** Consider the message `joker` encoded with block size $n=2$, cipher alphabet
`vowels` ($k=5$), and
$$A = \begin{pmatrix} 2 & 1 \\ 1 & 1 \end{pmatrix},
\qquad \mathbf{b} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}.$$

Since only characters in `aeiou` are encoded, the encoded string has the form  
`Why s_ s_r___s?! L_t's p_t _ sm_l_ _n th_t f_c_.`

The first block consists of the first two vowels: `o` (index 3) and `e` (index 1).
Applying the cipher: $A[3, 1]^\top + \mathbf{b} = [8, 4]^\top \equiv [3, 4] \pmod{5}$,
giving `o`, `u`. So the encoded string begins `Why so sur...`.

**Incomplete last block.** If the number of cipher-alphabet characters in the message is
not a multiple of $n$, the last incomplete block should be left unchanged.
